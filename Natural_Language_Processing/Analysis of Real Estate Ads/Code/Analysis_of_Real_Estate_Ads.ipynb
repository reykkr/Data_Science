##STEP 1: "REAL ESTATE AD" PYTHON OBJECT

````python
class RealEstateAd():

  def __init__(self,id):
    self.id = id
    self.sentences = None
    self.locations = []
    self.nbBathrooms = None
    self.nbBedrooms = None
    self.price = None
    self.adjectives = []
    self.salesPerson = None
    self.isActive = None
    print("Real Estate Ad object n°",id,"instanciate...")

  def getId(self):
    return self.id

  def setContent(self,content):
    self.content = content

  def getContent(self):
    return self.content

  def setSentences(self,lstOfSentences):
    self.sentences = lstOfSentences

  def getSentences(self):
    return self.sentences


  #Todo : Complete the object

  def addOneLocation(self, location):
    self.locations.append(location)
  
  def getLocations(self):
    return self.locations

  def setNbBathrooms(self,number):
    self.nbBathrooms = number

  def getNbBathrooms(self):
    return self.nbBathrooms

  def setNbBedrooms(self,number):
    self.nbBedrooms = number

  def getNbBedrooms(self):
    return self.nbBedrooms

  def setPrice(self,price):
    self.price = price
  
  def getPrice(self):
    return self.price
  
  def addOneAdjective(self,adjective):
    self.adjectives.append(adjective)
  
  def getAdjectives(self):
    return self.adjectives

  def setSalesPerson(self,salesPerson):
    self.salesPerson = salesPerson
  
  def getSalesPerson(self):
    return self.salesPerson

  def setIsActive(self,status):
    self.isActive = status

  def getIsActive(self):
    return self.isActive
  
````

````python
#--- TODO : TEST ON ONE AD BEFORE GENERALIZATION ----
# (Select the one you want)

#Instanciate the standard NERTagger
st = StanfordNERTagger('/content/stanford-ner-2020-11-17/classifiers/english.all.3class.distsim.crf.ser.gz',
                       '/content/stanford-ner-2020-11-17/stanford-ner.jar',
                       encoding='utf-8')


#Select the first a real estate ad
oRealEstateAd = lstOfAds[1]

#Selection of a sentence with a location
sentence = oRealEstateAd.getSentences()[0]
print(sentence)

#Tokenize sentence
tokens = word_tokenize(sentence)
print(tokens)

#Get Name Entities
ner = st.tag(tokens)
print(ner)

#Extraction of tags typed as a location
locations = []
for entity in ner:
  if entity[1]=="LOCATION":
    locations.append(entity[0])
locations
````

####**Exercise 5**

In the previous step, we identified the tokens typed as a location
However, given the results obtained, a location can be split into several tokens which does not make sense.

The objective is therefore to reconstitute the locations if necessary.

Tips :
- If the distance between the index of each location token in the sentence is equal to 1, we can consider that it is the same location
- Use the where function of the numpy library to find the index of a word in a sentence
- All locations for a given sentence will be stored in an array.

Expected result:


```
# ['Cherry', 'Creek', 'State', 'Park'] = ['Cherry Creek State Park']
```



````python
#--- TODO : TEST ON ONE AD BEFORE GENERALIZATION ----
# (Select the one you want)

#Convert tokens array to numpy array
numpyTokens = np.array(tokens)
print(numpyTokens)

#Finding index of each location token
indexLocations = []
for location in locations:
  index = np.where(numpyTokens == location)[0][0]
  indexLocations.append([location, index])
indexLocations
print(indexLocations)


#Reconstitution of location
lstLocations = []

for i in range(0,len(indexLocations)):

  #If we are at the first occurence, we save the name of the location and is index
  if i==0:
    strLoc = indexLocations[i][0]
    lastIndex = indexLocations[i][1]
  
    #For only one location
    if len(indexLocations)==1:
      lstLocations.append(strLoc)

  #Else
  else :
    locName = indexLocations[i][0]
    locIndex = indexLocations[i][1]
    
    #We calculate the index position difference from the previous token
    #If the difference is equal to 1, we estimate that we can concatenate
    #with the last token
    if((locIndex-lastIndex)==1):
      strLoc+=" "+locName
      lastIndex = locIndex
    
    #Else: We save the restructured location
    #and we start another location
    else:
      lstLocations.append(strLoc)
      strLoc = locName
      lastIndex = locIndex

    #Becareful: If we are at the end of the loop, add the reconstructed location
    if(i==len(indexLocations)-1):
      lstLocations.append(strLoc)

print(lstLocations)
````

````python
#TODO: Write your code here

for inx  in range(0,len(lstOfAds)):

  oRealEstateAd = lstOfAds[inx]

  numSentence = 1
  for sentence in oRealEstateAd.getSentences():

    #1: Tokenize sentence
    tokens = word_tokenize(sentence)

    #2: Name Entity Recognition
    ner = st.tag(tokens)
    locations = []
    for entity in ner:
      if entity[1]=="LOCATION":
        locations.append(entity[0])

    #3: Get index of locations
    npTokens = np.array(tokens)
    indexLocations = []
    for location in locations:
      index = np.where(npTokens == location)[0][0]
      indexLocations.append([location, index])

    #4: Reconstruct location
    lstLocations = []
    for i in range(0,len(indexLocations)):
      if i==0:
        strLoc = indexLocations[i][0]
        lastIndex = indexLocations[i][1]
        
        #For only one location
        if len(indexLocations)==1:
          lstLocations.append(strLoc)

      else :
        locName = indexLocations[i][0]
        locIndex = indexLocations[i][1]

        if((locIndex-lastIndex)==1):
          strLoc+=" "+locName
          lastIndex = locIndex
        else:
          lstLocations.append(strLoc)
          strLoc = locName
          lastIndex = locIndex

        if(i==len(indexLocations)-1):
          lstLocations.append(strLoc)


    print("AD N° ",inx+1," Sentence ",numSentence, " Location(s) : ",lstLocations)
    #5: Save locations to the advertising object
    for loc in lstLocations:
      oRealEstateAd.addOneLocation(loc)

    numSentence += 1

  lstOfAds[inx]=oRealEstateAd
````

####**Exercise 7**

First, we need to create a learning corpus.

````python
#TODO: Write your code here
import os
os.mkdir("train")
````

````python
listOfRooms = ['bedroom','bathroom','bedrooms','bathrooms','ceilings','mantel','kitchen','closet','en-suite,''porch','cabinetry','garage']

fileName = "own_ner.txt"
f = open("train/"+fileName,"w")
f.close()

for oRealEstateAd in lstOfAds:

  for sentence in oRealEstateAd.getSentences():

    #Get tokens
    tokens = word_tokenize(sentence)

    #Save the tag of the token
    for token in tokens:
      if token in listOfRooms: 
        tag = token+' ROOM\n'
      else :
        tag = token+' O\n'

      f = open("train/"+fileName,"a")
      f.write(tag)
      f.close()

````
