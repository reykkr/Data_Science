Your goal will be to create a Linear Regression Model, train it on the data with the optimal parameters using a grid search, and then evaluate the model's capabilities on a test set.


### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data

```python
df = pd.read_csv('AMES_Final_DF.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.columns
```

**The label we are trying to predict is the SalePrice column. Separate out the data into X features and y labels**

```python
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']
```

**Use scikit-learn to split up X and y into a training set and test set.**

Since we will later be using a Grid Search strategy, set your test proportion to 10%. 

To get the same data split as the solutions notebook, you can specify **random_state = 101**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101) #10% for testing.
```

```python
print(f" Training: {len(X_train)}, Testing: {len(X_test)}")
```

**The dataset features has a variety of scales and units. For optimal regression performance, scale the X features.**

Take carefuly note of what to use for .fit() vs what to use for .transform()

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

**We will use an Elastic Net model. Create an instance of default ElasticNet model with scikit-learn**

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

**The Elastic Net model has two main parameters, alpha and the L1 ratio. Create a dictionary parameter grid of values for the ElasticNet.**

```python
param_grid = {'alpha':[0.1, 1, 5, 10, 50, 100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1],
             'normalize' : [True, False]}
```

**Using scikit-learn create a GridSearchCV object and run a grid search for the best parameters for your model based on your scaled training data.**

```python
from sklearn.model_selection import GridSearchCV
```

```python
# verbose number a personal preference
grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error', 
                          cv=5,
                          verbose=1)
```

```python
grid_model.fit(scaled_X_train,y_train)
```

**Display the best combination of parameters for your model**

```python
grid_model.best_params_
```

**Evaluate your model's performance on the unseen 10% scaled test set.**

```python
predictions = grid_model.predict(scaled_X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
